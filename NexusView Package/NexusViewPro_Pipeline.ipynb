## 📝 Stage 2 – Write All Source `.py` Files

Each file is written **verbatim** to mirror the original project structure.


````python
def write_file(path: str, content: str):
    """Write content to path (creates parents if needed)."""
    p = Path(path)
    p.parent.mkdir(parents=True, exist_ok=True)
    p.write_text(content, encoding='utf-8')
    print(f'✍️  Written : {path}')
````

### `src/NexusViewPro/site.py`


````python
SITE_PY = '''\
from IPython import display
import urllib.request
from NexusViewPro.custom_exception import InvalidURLException
from NexusViewPro.logger import logger



def is_valid(URL: str) -> bool:
    try:
        response_status = urllib.request.urlopen(URL).getcode()
        assert response_status == 200
        logger.debug(f"response_status: {response_status}")
        return True
    except Exception as e:
        logger.exception(e)
        return False


def render_site(URL: str, width: str = "100%", height: str = "600") -> str:
    try:
        if is_valid(URL):
            response = display.IFrame(src=URL, width=width, height=height)
            display.display(response)
            return "success"
        else:
            raise InvalidURLException
    except Exception as e:
        raise e
'''

write_file('NexusViewPro/src/NexusViewPro/site.py', SITE_PY)
````

### `src/NexusViewPro/youtube.py`


````python
YOUTUBE_PY = r'''
from IPython.display import HTML, display
import re
from NexusViewPro.custom_exception import InvalidURLException
from NexusViewPro.logger import logger

def render_youtube_video(url: str, width: int = 780, height: int = 440):
  try:
    regex = r"(?:v=|\/)([0-9A-Za-z_-]{11}).*"
    match = re.search(regex,url)

    if not match:
      raise InvalidURLException(f"Invalid YouTube URL: {url}")

    video_id = match.group(1)
    embed_url = f"https://www.youtube-nocookie.com/embed/{video_id}"

    iframe = f"""
        <iframe width="{width}" height="{height}"
        src="{embed_url}"
        title="YouTube video player"
        frameborder="0"
        allow="accelerometer; autoplay; clipboard-write; encrypted-media; gyroscope; picture-in-picture"
        referrerpolicy="strict-origin-when-cross-origin"
        allowfullscreen>
        </iframe>
        """

    display(HTML(iframe))
    logger.info(f"Successfully rendered YouTube video for URL: {url}")

    return "success"

  except Exception as e:
    raise e
'''

write_file('NexusViewPro/src/NexusViewPro/youtube.py', YOUTUBE_PY)
````

## 📦 Stage 4 – Install Dependencies

> This installs the package in **editable mode** (`-e .`) from the `NexusViewPro/` folder.


````python
import subprocess, sys

def pip_install(*args):
    cmd = [sys.executable, '-m', 'pip', 'install', '--quiet'] + list(args)
    result = subprocess.run(cmd, capture_output=True, text=True)
    if result.returncode != 0:
        print('STDERR:', result.stderr[-800:])
    return result.returncode

# Core runtime deps
print('Installing core dependencies...')
rc = pip_install('ensure==1.0.2', 'py-youtube==1.1.7', 'IPython')
print(f'  → returncode={rc}')

# Install the package itself in editable mode
print('Installing NexusViewPro in editable mode...')
rc2 = pip_install('-e', 'NexusViewPro/')
print(f'  → returncode={rc2}')

print('\n✅ Dependencies installed.')
````
